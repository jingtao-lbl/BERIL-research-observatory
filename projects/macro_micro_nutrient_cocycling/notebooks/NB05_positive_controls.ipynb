## Summary

- All 8 targets encode both P-acquisition and metal-handling genes
- B. diazoefficiens and S. meliloti carry nifH+nifD (known diazotrophs)
- S. meliloti has the broadest repertoire: all 4 metal families + 8/9 P families
- P. chlororaphis is the only positive control with a phenazine operon
- S. coelicolor (= S. anthocyanicus in GTDB R214) carries phoD + pst + copA/corA + phzF
